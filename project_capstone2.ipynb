UJI KORELASI

```
data.select_dtypes(include='object').columns
```

1. uji korelasi last_review dan neighbourhood

```
from scipy.stats import kruskal


groups = [group['last_review_days'].dropna() for wilayah, group in data.groupby('neighbourhood')]

ha = "neighbourhood memiliki pengaruh terhadap last_review"
ho = "neighbourhood tidak memiliki pengaruh terhadap last_review"

stat, p = kruskal(*groups)
if p < 0.05:
    print(f"{ha}")
else:
    print(f"{ho}")

# Hitung epsilon-squared 'seberapa berpengaruh'
k = len(groups)
n = sum(len(group) for group in groups)
epsilon_sq = (stat - k + 1) / (n - k)

print(f"Kruskal-Wallis H-statistic: {stat}, p-value: {p: .2e}")
print(f"Epsilon-squared effect size: {epsilon_sq:.4f}")
```

2. melihat korelasi last_review dan room_type

```
from scipy.stats import kruskal


groups = [group['last_review_days'].dropna() for wilayah, group in data.groupby('room_type')]

ha = "room_type memiliki pengaruh terhadap last_review_days"
ho = "room_type tidak memiliki pengaruh terhadap last_review_days"

stat, p = kruskal(*groups)
if p < 0.05:
    print(f"{ha}")
else:
    print(f"{ho}")

# Hitung epsilon-squared 'seberapa berpengaruh'
k = len(groups)
n = sum(len(group) for group in groups)
epsilon_sq = (stat - k + 1) / (n - k)

print(f"Kruskal-Wallis H-statistic: {stat}, p-value: {p: .2e}")
print(f"Epsilon-squared effect size: {epsilon_sq:.4f}")
```

3. uji statistika korelasi numerik

```
from scipy.stats import spearmanr
subset = data[['last_review_days', 'reviews_per_month']].dropna()
stat, p_value = spearmanr(subset['last_review_days'], subset['reviews_per_month'])

ha = "last_review_days memiliki korelasi yang kuat dengan reviews_per_month"
ho = "last_review_days tidak memiliki korelasi yang kuat dengan reviews_per_month"

if p_value < 0.05:
    print(f"{ha}")
else:
    print(f"{ho}")

print(f"Spearman correlation: {stat}, p-value: {p_value}")
```

### Uji pengaruh price dengan neighbourhood

```
from scipy.stats import kruskal


groups = [group['price'] for wilayah, group in data.groupby('neighbourhood')]

ha = "neighbourhood memiliki pengaruh terhadap price"
ho = "neighbourhood tidak memiliki pengaruh terhadap price"

stat, p = kruskal(*groups)
if p < 0.05:
    print(f"{ha}")
else:
    print(f"{ho}")

# Hitung epsilon-squared 'seberapa berpengaruh'
k = len(groups)
n = sum(len(group) for group in groups)
epsilon_sq = (stat - k + 1) / (n - k)

print(f"Kruskal-Wallis H-statistic: {stat}, p-value: {p: .2e}")
print(f"Epsilon-squared effect size: {epsilon_sq:.4f}")
```

### uji korelasi neighbourhood dengan room_type

```
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

# Buat tabel kontingensi
table = pd.crosstab(data['neighbourhood'], data['room_type'])

# Uji chi-square
chi2, p, dof, expected = chi2_contingency(table)

# Cramér’s V
n = table.sum().sum()
phi2 = chi2 / n
r, k = table.shape
cramers_v = np.sqrt(phi2 / min(k-1, r-1))

ha = "room_type dan neighbourhood memiliki korelasi"
ho = "room_type dan neighbourhood tidak memiliki korelasi"

if p < 0.05:
    print(f"{ha}")
else:
    print(f"{ho}")

print(f"Chi2 = {chi2:.2f}, p = {p:.4f}")
print(f"Cramér's V = {cramers_v:.4f}")
```

### uji korelasi room_type dengan price

```
from scipy.stats import kruskal


groups = [group['price'] for wilayah, group in data.groupby('room_type')]

ha = "room_type memiliki pengaruh terhadap price"
ho = "room_type tidak memiliki pengaruh terhadap price"

stat, p = kruskal(*groups)
if p < 0.05:
    print(f"{ha}")
else:
    print(f"{ho}")

# Hitung epsilon-squared 'seberapa berpengaruh'
k = len(groups)
n = sum(len(group) for group in groups)
epsilon_sq = (stat - k + 1) / (n - k)

print(f"Kruskal-Wallis H-statistic: {stat}, p-value: {p: .2e}")
print(f"Epsilon-squared effect size: {epsilon_sq:.4f}")
```

*   id : Mau tahu listing seperti apa yang laku di pasaran
*   host_id : Mau tahu siapa host terbaik



```
rfm = data.groupby('id').agg({
    'last_review': lambda x: (today - x.max()).days, # R
    'number_of_reviews': 'sum', # F
    'price': 'mean' # M
}).reset_index()
# rename
rfm.columns = ['id', 'Recency', 'Frequency', 'Monetary']
rfm
```

### 4.Konversi RFM ke Skor 1–5

```
# Skor 1–5: semakin kecil Recency (lebih baru), semakin tinggi skornya
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

### 5.buat pivot table untuk grafik heatmap

```
# Gunakan R dan F sebagai sumbu, dan nilai rata-rata M sebagai isi
rfm_pivot = rfm.pivot_table(index='F_Score', columns='R_Score', values='Monetary', aggfunc='mean')
```

### plot heatmap

```
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches  # jangan lupa ini

rfm_pivot = rfm.pivot_table(
    index='F_Score', columns='R_Score', values='Monetary', aggfunc='mean'
)

plt.figure(figsize=(15,6))
sns.heatmap(rfm_pivot, annot=True, fmt=".1f", cmap='RdYlGn')
plt.title("Heatmap RFM - Monetary per Recency & Frequency Score")
plt.xlabel("Recency Score : last_review")
plt.ylabel("Frequency Score : number_of_reviews")

red_patch = mpatches.Patch(color='red', label='Merah: Nilai Rendah (Listing Kurang Laku / Harga Murah) score : 1')
yellow_patch = mpatches.Patch(color='yellow', label='Kuning: Nilai Sedang (Listing Potensial) score : 3')
green_patch = mpatches.Patch(color='green', label='Hijau: Nilai Tinggi (Listing Laku / Harga Tinggi) score : 5')

plt.legend(handles=[red_patch, yellow_patch, green_patch], bbox_to_anchor=(1.2, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
